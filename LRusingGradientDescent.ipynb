```
x = np.array(df.iloc[:, 0]).reshape(-1,1)
```

```
y = np.array(df.iloc[:, 2]).reshape(-1,1)
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```
plt.plot(x, y, "r*--")
```

```
def gradient_descent(x,y):
  m = 0
  b = 0
  iterations = 15000
  lr = 0.0016857
  n = len(x)
  


  for i in range(iterations):
   
    y_guess = m * x + b
    mse = (1/n) * sum([v**2 for v in (y-y_guess)])
    m_derivative = -(2/n) * sum(x*(y-y_guess))
    b_derivative = -(2/n) * sum(y-y_guess)
    m = m - lr * m_derivative
    b = b - lr * b_derivative
    r2 = r2_score(y_train, y_guess)
    print("m = {}, b = {}, iteration = {}, mean squared error = {}, r2 = {}".format(m, b, i, mse, r2))
  
```

```
gradient_descent(x_train, y_train)
```
